```python
global index_
index_ = [(22,'Черкаська'),(24,'Чернігівська'),(23,'Чернівецька'),(25, 'Республіка Крим'),(3, 'Дніпропетровська'),(4,'Донецька'), (8, 'Івано-Франківська'),(19, 'Харківська'),(20, 'Херсонська'),(21, 'Хмельницька'),(9,'Київська'),(26,'Київ'),(10,'Кіровоградська'),(11,'Луганська'),(12,'Львівська'),(13,'Миколаївська'),(14,'Одеська'),(15,'Полтавська'),(16,'Рівенська'),(27,'Севастополь'),(17,'Сумська'),(18,'Тернопільська'),(6,'Закарпатська'),(1,'Вінницька'),(2,'Волинська'),(7,'Запорізька'),(5,'Житомирська')]
index_ = list(enumerate(index_, start = 1))
print (index_)
```

```python
def download_data(index):
    url = f"https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_TS_admin.php?country=UKR&provinceID={index}&year1=1981&year2=2020&type=Mean"
    mas = requests.get(url)
    data = ""
    _, data = mas.text.split('<br><tt><pre>')
    data = data.replace('</pre></tt>', '')
    if int(index_[int(index)-1][1][0]) < 10:
        filename = f"0-{index_[int(index)-1][1][0]}-1981-2002.csv"
    elif (10 <= int(index_[int(index)-1][1][0])) and (int(index_[int(index)-1][1][0]) < 20):
        filename = f"1-{str(int(index_[int(index)-1][1][0])-10)}-1981-2002.csv"
    elif (20 <= int(index_[int(index)-1][1][0])):
        filename = f"2-{str(int(index_[int(index)-1][1][0])-20)}-1981-2002.csv"
    #filename = f"{index_[int(index)-1][1][0]}-1981-2002.csv"
    with open(filename, 'w') as f:
        f.write(data)

    return 0
```

```python
def download_all_data():
    for i in range(0, 27):
        download_data(str(index_[i][0]))
        print(f"Інформація {index_[i][1][1]} завантажена")
        
download_all_data()
```

```python
def csv_directory(directory):
    df = pd.DataFrame()

    for file in os.listdir(directory):
        
        if file.endswith('.csv'):
            region1,region2, *_= file.split('-')
            dt = pd.read_csv(file, header=None)
            dt.rename(columns={0: 'рік', 1: 'тиждень', 2: 'NDVI', 3: 'BT', 4: 'VCI', 5: 'TCI', 6: 'VHI'}, inplace=True)
            for i in range(0, 27):
                if int(index_[i][1][0]) == (int(region1)*10 + int(region2)):
                   dt['область'] = index_[i][1][1] 
            dt = dt.drop(7, axis=1)
            df = df.append(dt, ignore_index=True)
    return df

df = csv_directory('.')
without_index = df[df['NDVI'] == -1].index
df = df.drop(without_index)
df = df.set_index(['область', 'рік', 'тиждень'])
df
```
